# Linear Regression Algo uza OLS (Ordinary least Square)

```python
import pandas as pd
import numpy as np

# Create a simple dataset for linear regression
deviation = 5 # tweek ida ne'e atu muda deviation husi Dependent variable (y)
X = np.arange(1, 31) # tweek ida ne'e atu muda size husi dataest exemplo iah ne'e 1-30
y = 3 * X + np.random.normal(0, deviation, size=X.shape) #
data = pd.DataFrame({'X': X, 'y': y})
data.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(data['X'], data['y'])
```

```python
data_calc = data.copy()
```

```python




x_mean, y_mean = data_calc['X'].mean(), data_calc['y'].mean() # buka nia mean

# buka kada devaition husi data points husi mean
data_calc['X_minus_meanX'] = data_calc['X'] - x_mean
data_calc['y_minus_meanY'] = data_calc['y'] - y_mean

# calcula deviationX * DeviationY
data_calc['deviationXY'] = data_calc['X_minus_meanX'] * data_calc['y_minus_meanY']

sum_deviationXY = data_calc['deviationXY'].sum()

sum_deeviationX2 = (data_calc['X_minus_meanX'] ** 2).sum()

B1 = sum_deviationXY / sum_deeviationX2
B1

B0 = y_mean-B1*x_mean
print(f"B0: {B0}, B1: {B1}")
print(f"y = {B0} + {B1}X")
```

```python
plt.scatter(data['X'], data['y'])
y_pred = B0 + B1*data['X']
plt.plot(data['X'], y_pred, color='red')
```

# LOSS FUNCTION

mak hanesan function ida ne'e uza atu ukur *how well a line fit it's data points*. Loss function ne'ebe babain uza mak Mean Squared Error

nia formula maka 

**Residual Error**
$$
e_i = y_i - \hat{y}_i
$$​

**MSE**
$$
MSE = \frac{1}{n} \sum_{i=1}^n (y_i - \hat{y}_i)^2
$$


## codidng implementation


```python
mse = ((data['y'] - y_pred) ** 2).mean()
print("Mean Squared Error:", mse)
```
